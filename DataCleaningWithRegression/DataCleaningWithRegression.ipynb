```python
# Veri setini okuma
import pandas as pd

df = pd.read_csv('titles_1.csv')
```

```python
# Veri setinin temel bilgilerini görüntülüyoruz
df.info()
```

```python
# Her sütundaki eksik değerlerin toplam sayısını hesaplıyoruz
df.isnull().sum()
```

```python
# Tekrarlanan satır sayısını hesaplıyoruz
df.duplicated().sum()
```

```python
# Veri setindeki gereksiz sütunları çıkararak yeni bir DataFrame oluşturuyoruz
data = df.drop("title", axis=1)
data.head()
```

```python
# "id" sütununu çıkarıyoruz
data.drop("id",axis=1, inplace=True)
data.head()
```

```python
# "imdb_id" sütununu çıkarıyoruz
data.drop("imdb_id", axis=1, inplace=True)
data.head()
```

```python
# "description" sütununu çıkarıyoruz
data.drop("description", axis=1, inplace=True)
data.head()
```

```python
# "age_certification" sütunundaki eksik değerleri "Not Rated" ile dolduruyoruz
data["age_certification"].fillna("Not Rated", inplace = True)
data.head()
```

```python
# "seasons" sütunundaki eksik değerleri 0 ile dolduruyoruz
data["seasons"].fillna(0, inplace = True)
data.head()
```

```python
# önceki işlemlerden sonra kontrol
data.isnull().sum()
```

```python
# "imdb_score" sütunundaki eksik değerleri ortalama IMDb puanı ile dolduruyoruz
imdbScore = data["imdb_score"].mean()
data["imdb_score"].fillna(imdbScore, inplace=True)
data.head()
```

```python
# "imdb_votes" sütunundaki eksik değerleri ortalama IMDb oy sayısı ile dolduruyoruz
imdbVotes = data["imdb_votes"].mean()
data["imdb_votes"].fillna(imdbVotes, inplace = True)
data.head()
```

```python
# "tmdb_popularity" sütunundaki eksik değerleri ortalama TMDB popülerlik puanı ile dolduruyoruz
tmdbPop = data['tmdb_popularity'].mean()
data['tmdb_popularity'].fillna(tmdbPop, inplace=True)
data.head()
```

```python
# "tmdb_score" sütunundaki eksik değerleri ortalama TMDB puanı ile dolduruyoruz
tmdbScore = data["tmdb_score"].mean()
data["tmdb_score"].fillna(tmdbScore, inplace=True)
data.head()
```

```python
# kontrol
data.isnull().sum()
```

```python
# LabelEncoder kullanarak kategorik sütunları sayısal değerlere dönüştürüyoruz
from sklearn.preprocessing import LabelEncoder

dataYedek = data.copy() # Veri setini yedekliyoruz

lblencoder = LabelEncoder()

data["age_certification"] = lblencoder.fit_transform(data["age_certification"])
data["type"] = lblencoder.fit_transform(data["type"])
data["genres"] = lblencoder.fit_transform(data["genres"])
data["production_countries"] = lblencoder.fit_transform(data["production_countries"])

data.head()
```

```python
# Korelasyon matrisi görselleştirmesi
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.figure(figsize=(10, 10))
corr = data.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, cmap="coolwarm", mask=mask, xticklabels=corr.columns, yticklabels=corr.columns, fmt=".2f", annot_kws={"size": 10}, linewidths=0.5, square=True)
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
# Yaş sınıflandırması ve tür arasındaki ilişkiyi gösteren grafik
Grafik = dataYedek[["age_certification", "type"]]

plt.figure(figsize=(10, 6))
sns.countplot(data=Grafik, x="age_certification", hue="type", palette="hot")
plt.title("Yaş Sınıflandırması ve Tip İlişkisi")
plt.xlabel("Yaş Sınıflandırması")
plt.ylabel("Sayı")
plt.legend(title="Tür")
plt.show()
```

```python
# imdb_score ve tmdb_score arasındaki ilişkiyi gösteren grafik
Grafik0 = dataYedek[["tmdb_score", "imdb_score", "type"]]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=Grafik0, x="tmdb_score", y="imdb_score", hue="type", palette="cool")
plt.title("TMDB Score vs IMDB Score")
plt.xlabel("TMDB Score")
plt.ylabel("IMDB Score")
plt.legend(title="Tür")
plt.show()
```

```python
# imdb_score sütununu ayırma
x = data.drop("imdb_score", axis=1)
y = data["imdb_score"]
```

```python
# Train test split işlemleri
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.25)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor

knnReg = KNeighborsRegressor()
knnReg.fit(x_train, y_train)
predKNN = knnReg.predict(x_test)
knnMSE = mean_squared_error(y_test, predKNN)

rfReg = RandomForestRegressor(random_state=42)
rfReg.fit(x_train, y_train)
predRF = rfReg.predict(x_test)
rfMSE = mean_squared_error(y_test, predRF)

gbReg = GradientBoostingRegressor(random_state=42)
gbReg.fit(x_train, y_train)
predGB = gbReg.predict(x_test)
gbMSE = mean_squared_error(y_test, predGB)
```

```python
# Hiperparametreler
knnParams = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'], 
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
}

rfParams = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

gbParams = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Gridsearch kullanma
from sklearn.model_selection import GridSearchCV

knnGridsearch = GridSearchCV(estimator=KNeighborsRegressor(), param_grid=knnParams, cv=5)
knnGridsearch.fit(x_train, y_train)
bestKNN = knnGridsearch.best_estimator_
```

```python
rfGridsearch = GridSearchCV(estimator=RandomForestRegressor(random_state=42), param_grid=rfParams, cv=5)
rfGridsearch.fit(x_train, y_train)
bestRF = rfGridsearch.best_estimator_
```

```python
gbGridsearch = GridSearchCV(estimator=GradientBoostingRegressor(random_state=42), param_grid=gbParams, cv=5)
gbGridsearch.fit(x_train, y_train)
bestGB = gbGridsearch.best_estimator_
```

```python
# En iyi parametreleri ve MSE değerlerini yazdırma
print("KNN en iyi parametreler:", bestKNN.get_params())
print("KNN mse:", bestKNN)
print("\n")

print("Random forest en iyi parametreler:", bestRF.get_params())
print("Random forest MSE:", bestRF)
print("\n")

print("Gradient boosting en iyi parametreler:", bestGB.get_params())
print("Gradient boosting MSE:", bestGB)
```

```python
# Grafik oluşturma
EniyiMSE = [knnGridsearch.best_score_, rfGridsearch.best_score_, gbGridsearch.best_score_]
defaultMSE = [knnMSE, rfMSE, gbMSE]
algorithms = ['KNN', 'Random Forest', 'Gradient Boosting']

# Grafiği oluşturma
plt.figure(figsize=(10, 6))
plt.plot(algorithms, EniyiMSE, marker='o', label='Best Parameters')
plt.plot(algorithms, defaultMSE, marker='x', label='Default Parameters')
plt.xlabel('Algorithms')
plt.ylabel('MSE')
plt.title('Default ve En iyi Parametrelerin MSE Karşılaştırması')
plt.legend()
plt.grid(True)
plt.show()
```
